```python
import pandas as pd
```

```python
#importing data
df=pd.read_csv("HR-Employee-Attrition.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
#convert Y (Attrition) into numbers
df["Attrition"].replace(to_replace='Yes', value=1, inplace=True)
df["Attrition"].replace(to_replace='No', value=0, inplace=True)
df.head()
```

```python
#convert gender into numbers
df["Gender"].replace(to_replace='Female', value=1, inplace=True)
df["Gender"].replace(to_replace='Male', value=0, inplace=True)
df["Gender"].head()
```

```python
# check the total  number of missing  value 
df.isnull().sum()
```

```python
#removing columns that don't lead to employee attrition 
df=df.drop('EmployeeCount',axis=1)
df=df.drop('StandardHours',axis=1)
df=df.drop('Over18',axis=1)
```

```python
#Let's convert all the categorical variables into dummy variables
df_dummies = pd.get_dummies(df)

# show  the first 5 rows 
df_dummies.head()
```

```python
df_dummies.info()
```

Factors that lead to employee attrition

```python
#Get Correlation of "Attirition" with other variables:
df_dummies.corr()['Attrition'].sort_values(ascending = False)
```

Overtime_Yes, MaritalStatus_Single, JobRole_Sales Representative and BusinessTravel_Travel_Frequently seem to be the top positively corellated factors with Attrition
                  
OverTime_No, TotalWorkingYears, JobLevel , YearsInCurrentRole seem to be the top negatively corellated factors with Attrition


__A breakdown of distance from home by job role and attrition.__

```python
distancefrhome_JobRole_attrition = df[["DistanceFromHome", "Attrition", "JobRole"]]
distancefrhome_JobRole_attrition= distancefrhome_JobRole_attrition.groupby(["JobRole", "Attrition"]).agg('mean')
print(distancefrhome_JobRole_attrition)
distancefrhome_JobRole_attrition.unstack(level=-1)['DistanceFromHome']
```

__Average monthly income by education and attrition.__

```python
edu_attrition_income = df[["Education", "Attrition", "MonthlyIncome"]]
edu_attrition_income = edu_attrition_income.groupby(["Education", "Attrition"]).agg('mean')
print(edu_attrition_income)
edu_attrition_income.unstack(level=-1)['MonthlyIncome']
```
